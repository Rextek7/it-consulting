## Machine learning

```python
def compute_metrics(y_true, y_pred, m=1):
    """
    Compute forecast accuracy metrics: WAPE, MAE, RMSE, MASE.

    Parameters
    ----------
    y_true : array-like of shape (n,)
        Истинные значения.
    y_pred : array-like of shape (n,)
        Предсказанные значения.
    m : int, default=1
        Шаг «наивного» прогноза для расчёта MASE.
        Для несезонного ряда обычно m=1;
        для сезонного — период сезонности.

    Returns
    -------
    dict
        {
            'WAPE': float,  # в процентах
            'MAE': float,
            'RMSE': float,
            'MASE': float
        }
    """
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    n = len(y_true)

    # WAPE
    num = np.abs(y_true - y_pred).sum()
    denom = np.abs(y_true).sum()
    wape = num / denom * 100 if denom != 0 else np.nan

    # MAE
    mae = np.mean(np.abs(y_true - y_pred))

    # RMSE
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

    # MASE
    if n > m:
        naive_errors = np.abs(y_true[m:] - y_true[:-m])
        scale = np.mean(naive_errors)
        mase = mae / scale if scale != 0 else np.nan
    else:
        mase = np.nan

    return {
        'WAPE': wape,
        'MAE': mae,
        'RMSE': rmse,
        'MASE': mase
    }
```

### Загрузка данных для группы 18 — "Рис басмати 500 гр"

- Загружаются обучающая и тестовая выборки, относящиеся к 18-й группе товара.
- Обучающая выборка фильтруется по периоду: с 5 февраля 2021 года по 17 января 2022 года.
- В тестовой выборке столбец `BPV` переименовывается в `BPV_true` для единообразия и дальнейшего сравнения с предсказаниями модели.

```python
train_path = "data_grouped/train_group_18_Рис басмати 500 гр.csv"
test_path = "data_test_grouped/test_group_18_Рис басмати 500 гр.csv"

train = pd.read_csv(train_path)
train = train[(train["Period"] >= "2021-02-05") & (train["Period"] <= "2022-01-17")].copy()
test = pd.read_csv(test_path).rename(columns={"BPV": "BPV_true"})
```

### Функция `enrich`: генерация временных и сезонных признаков

Функция добавляет в датафрейм `df` следующие признаки, отражающие временные закономерности:

- `Period` — преобразование в формат `datetime`.
- `Period_ord` — числовое представление даты (количество дней от начала эпохи).
- `weekofyear` — номер недели в году.
- `month`, `day`, `days_in_month` — календарные признаки, включая количество дней в месяце.
- `week_share` — доля месяца, пройденная на текущую дату.
- `sin_month`, `cos_month` — синус и косинус месяца, моделируют циклический сезонный эффект.
- `Period_last_year` — дата, соответствующая текущей неделе год назад (смещение на 52 недели).

Эти признаки используются для улучшения качества прогноза за счёт учёта календарных и сезонных закономерностей.

```python
def enrich(df):
    df["Period"] = pd.to_datetime(df["Period"])
    df["Period_ord"] = df["Period"].map(pd.Timestamp.toordinal)
    df["weekofyear"] = df["Period"].dt.isocalendar().week.astype(int)
    df["month"] = df["Period"].dt.month
    df["day"] = df["Period"].dt.day
    df["days_in_month"] = df["Period"].dt.days_in_month
    df["week_share"] = df["day"] / df["days_in_month"]
    df["sin_month"] = np.sin(2 * np.pi * df["month"] / 12)
    df["cos_month"] = np.cos(2 * np.pi * df["month"] / 12)
    df["Period_last_year"] = df["Period"] - pd.DateOffset(weeks=52)
    return df
```

### Обогащение выборок временными признаками

Применение функции `enrich` к обучающей (`train`) и тестовой (`test`) выборкам позволяет добавить набор временных и сезонных признаков.
Это необходимо для того, чтобы модель могла учитывать закономерности, связанные с календарём и сезонностью при прогнозировании спроса.

```python
train = enrich(train)
test = enrich(test)
```

### Добавление признака продаж за аналогичный период прошлого года

- Создаётся таблица `sellin_map`, в которой:
  - `Period` переименовывается в `Period_last_year`,
  - `Total Sell-in` — в `sellin_last_year`.
- Выполняется объединение обучающей и тестовой выборок с этой таблицей по `Period_last_year`,
  что позволяет добавить продажи за аналогичную неделю предыдущего года.
- Пропущенные значения (если данных за прошлый год не было) заменяются нулями.
- Добавляется логарифмированный признак `log_sellin_last_year` — он сглаживает масштаб продаж и уменьшает влияние выбросов.

```python
sellin_map = train[["Period", "Total Sell-in"]].rename(
    columns={"Period": "Period_last_year", "Total Sell-in": "sellin_last_year"}
)
train = train.merge(sellin_map, on="Period_last_year", how="left")
test = test.merge(sellin_map, on="Period_last_year", how="left")
train["sellin_last_year"].fillna(0, inplace=True)
test["sellin_last_year"].fillna(0, inplace=True)

train["log_sellin_last_year"] = np.log1p(train["sellin_last_year"])
test["log_sellin_last_year"] = np.log1p(test["sellin_last_year"])
```

### Подготовка данных для классификации и регрессии

- `feat_cols` — список признаков, включающих календарные, сезонные и отложенные характеристики, которые будут использоваться для обучения моделей.
- `train_clf` — подмножество обучающей выборки с известными значениями `BPV`.
- `sale_flag` — бинарная целевая переменная: 1, если `BPV > 0`, иначе 0. Используется для задачи классификации факта продаж.
- `X_clf`, `y_clf` — признаки и целевая переменная для обучения классификатора.
- `train_reg` — данные только с положительными значениями `BPV`, предназначенные для обучения регрессионной модели.
- `X_reg`, `y_reg` — признаки и логарифмированные объёмы продаж для регрессии.
- `X_test` — матрица признаков из тестовой выборки, по которой будут сделаны предсказания.
- `true_test` — таблица с датами и фактическими значениями `BPV_true`, используемая для оценки качества прогноза.

```python
feat_cols = [
    "Period_ord", "weekofyear", "month",
    "sellin_last_year", "log_sellin_last_year",
    "week_share", "sin_month", "cos_month"
]

train_clf = train[train["BPV"].notna()].copy()
train_clf["sale_flag"] = (train_clf["BPV"] > 0).astype(int)

X_clf = train_clf[feat_cols]
y_clf = train_clf["sale_flag"]

train_reg = train_clf[train_clf["sale_flag"] == 1].copy()
X_reg = train_reg[feat_cols]
y_reg = np.log1p(train_reg["BPV"])

X_test = test[feat_cols].reset_index(drop=True)
true_test = test[["Period", "BPV_true"]].reset_index(drop=True)
```

### Обучение ансамбля XGBoost-моделей для классификации и регрессии

Цикл запускает ансамбль из 5 моделей с разными `random_state` для повышения устойчивости предсказаний:

1. **Классификатор (`XGBClassifier`)**:
   - Обучается на задаче бинарной классификации: предсказать, будет ли продажа в периоде (`sale_flag`).
   - Использует:
     - очень низкий `learning_rate=0.0005` — предполагается осторожное обучение,
     - `scale_pos_weight` — для балансировки классов при дисбалансе между нулями и единицами,
     - `logloss` как метрику качества.
   - Предсказанная вероятность продажи сохраняется в `probas`.

2. **Регрессор (`XGBRegressor`)**:
   - Обучается на примерах с положительными продажами (логарифмированными).
   - Предсказывает значения, затем преобразует их обратно с помощью `np.expm1`.
   - Отрицательные значения обнуляются (продажи не могут быть отрицательными).
   - Предсказания сохраняются в список `preds`.

Итог: создаются два ансамбля — один для вероятности факта продаж, второй для объёма продаж, если они состоятся.

```python
N_MODELS = 5
probas = []
preds = []

for seed in range(42, 42 + N_MODELS):
    clf = XGBClassifier(
        n_estimators=500,
        learning_rate=0.0005,
        max_depth=5,
        scale_pos_weight=(y_clf == 0).sum() / (y_clf == 1).sum(),
        use_label_encoder=False,
        eval_metric="logloss",
        random_state=seed,
        verbosity=0
    )
    clf.fit(X_clf, y_clf)
    proba_test = clf.predict_proba(X_test)[:, 1]
    probas.append(proba_test)

    reg = XGBRegressor(
        n_estimators=600,
        learning_rate=0.05,
        max_depth=5,
        random_state=seed,
        verbosity=0
    )
    reg.fit(X_reg, y_reg)
    pred = np.expm1(reg.predict(X_test))
    preds.append(np.maximum(pred, 0))
```

### Агрегация результатов ансамбля и оценка модели

- **Агрегация предсказаний:**
  - `proba_test_median` вычисляет медиану вероятностей продажи по всем классификаторам.
  - `base_pred_median` вычисляет медианное предсказание объёма продаж по всем регрессорам.
- **Принятие решения о наличии продаж:**
  - Используется порог `threshold = 0.1`. Если медианная вероятность продажи больше или равна порогу, считается, что продажа состоится, и применяется медианное предсказание объёма; иначе прогнозируется 0.
- **Формирование итогового датафрейма:**
  - `merged` содержит даты (`Period`), фактические продажи (`BPV_true`) и предсказанные значения (`BPV_pred`).
- **Оценка качества модели:**
  - Вычисляются метрики WAPE, MAE, RMSE и MASE (при параметре сезонности `m=1`).
  - Результаты метрик выводятся, что позволяет оценить качество предсказаний.

```python
proba_test_median = np.median(probas, axis=0)
base_pred_median = np.median(preds, axis=0)

threshold = 0.1
is_sale = proba_test_median >= threshold
y_pred = np.where(is_sale, base_pred_median, 0)

merged = pd.DataFrame({
    "Period": true_test["Period"],
    "BPV_true": true_test["BPV_true"],
    "BPV_pred": y_pred
})

metrics = compute_metrics(merged["BPV_true"], merged["BPV_pred"], m=1)

print(f"WAPE = {metrics['WAPE']:.2f}%")
print(f"MAE  = {metrics['MAE']:.4f}")
print(f"RMSE = {metrics['RMSE']:.4f}")
print(f"MASE = {metrics['MASE']:.4f}")
```

- Из итогового датафрейма `merged` удаляется колонка `BPV_true`, чтобы оставить только даты и предсказанные значения `BPV_pred`.
