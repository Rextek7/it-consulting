## Machine learning

```python
def compute_metrics(y_true, y_pred, m=1):
    """
    Compute forecast accuracy metrics: WAPE, MAE, RMSE, MASE.

    Parameters
    ----------
    y_true : array-like of shape (n,)
        Истинные значения.
    y_pred : array-like of shape (n,)
        Предсказанные значения.
    m : int, default=1
        Шаг «наивного» прогноза для расчёта MASE.
        Для несезонного ряда обычно m=1;
        для сезонного — период сезонности.

    Returns
    -------
    dict
        {
            'WAPE': float,  # в процентах
            'MAE': float,
            'RMSE': float,
            'MASE': float
        }
    """
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    n = len(y_true)

    # WAPE
    num = np.abs(y_true - y_pred).sum()
    denom = np.abs(y_true).sum()
    wape = num / denom * 100 if denom != 0 else np.nan

    # MAE
    mae = np.mean(np.abs(y_true - y_pred))

    # RMSE
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

    # MASE
    if n > m:
        naive_errors = np.abs(y_true[m:] - y_true[:-m])
        scale = np.mean(naive_errors)
        mase = mae / scale if scale != 0 else np.nan
    else:
        mase = np.nan

    return {
        'WAPE': wape,
        'MAE': mae,
        'RMSE': rmse,
        'MASE': mase
    }
```

### Загрузка данных для товара "Рис басмати 500 гр"

- Загружаются обучающая и тестовая выборки для группы 2 — "Рис басмати 500 гр".
- Обучающая выборка фильтруется по периоду с 20 сентября 2019 года по 17 января 2022 года.
- В тестовой выборке столбец `BPV` переименовывается в `BPV_true` для унификации структуры и последующего сравнения с предсказаниями.

```python
train_path = "data_grouped/train_group_2_Рис басмати 500 гр.csv"
test_path = "data_test_grouped/test_group_2_Рис басмати 500 гр.csv"

train = pd.read_csv(train_path)
train = train[(train["Period"] >= "2019-09-20") & (train["Period"] <= "2022-01-17")].copy()
test = pd.read_csv(test_path).rename(columns={"BPV": "BPV_true"})
```

### Функция `enrich`: генерация временных признаков

Функция `enrich(df)` добавляет к датафрейму набор временных признаков, полезных для моделирования сезонности и календарных эффектов:

- `Period` — преобразование в формат `datetime`.
- `Period_ord` — порядковый номер даты (числовой формат), используемый как временной признак.
- `weekofyear` — номер недели в году.
- `month`, `day` — месяц и день месяца.
- `days_in_month` — количество дней в текущем месяце.
- `week_share` — доля месяца, прошедшая к текущей дате (`day` / `days_in_month`).
- `sin_month`, `cos_month` — синус и косинус месяца; используются для моделирования циклической сезонности.
- `Period_last_year` — дата, соответствующая аналогичному периоду год назад (сдвиг на 52 недели назад).

```python
def enrich(df):
    df["Period"] = pd.to_datetime(df["Period"])
    df["Period_ord"] = df["Period"].map(pd.Timestamp.toordinal)
    df["weekofyear"] = df["Period"].dt.isocalendar().week.astype(int)
    df["month"] = df["Period"].dt.month
    df["day"] = df["Period"].dt.day
    df["days_in_month"] = df["Period"].dt.days_in_month
    df["week_share"] = df["day"] / df["days_in_month"]
    df["sin_month"] = np.sin(2 * np.pi * df["month"] / 12)
    df["cos_month"] = np.cos(2 * np.pi * df["month"] / 12)
    df["Period_last_year"] = df["Period"] - pd.DateOffset(weeks=52)

    return df
```

### Обогащение обучающей и тестовой выборок временными признаками

Функция `enrich` применяется к обеим выборкам — `train` и `test`, добавляя в них календарные и сезонные признаки (дата в числовом виде, номер недели, месяц, синус/косинус месяца и т.д.).
Это позволяет модели учитывать временные закономерности при обучении и прогнозировании.

```python
train = enrich(train)
test = enrich(test)
```

### Добавление информации о продажах за аналогичный период прошлого года

- Формируется таблица `sellin_map`, содержащая продажи (`Total Sell-in`) и соответствующие им даты, переименованные в `Period_last_year` и `sellin_last_year`.
- Объединение с `train` и `test` по признаку `Period_last_year` позволяет добавить информацию о продажах в ту же неделю предыдущего года.
- Пропущенные значения заполняются нулями, что устраняет потенциальные пропуски при отсутствии данных за прошлый год.
- Создаётся логарифмированный признак `log_sellin_last_year` — сглаживает масштаб продаж и уменьшает влияние выбросов.

```python
sellin_map = train[["Period", "Total Sell-in"]].rename(
    columns={"Period": "Period_last_year", "Total Sell-in": "sellin_last_year"}
)
train = train.merge(sellin_map, on="Period_last_year", how="left")
test = test.merge(sellin_map, on="Period_last_year", how="left")
train["sellin_last_year"].fillna(0, inplace=True)
test["sellin_last_year"].fillna(0, inplace=True)

train["log_sellin_last_year"] = np.log1p(train["sellin_last_year"])
test["log_sellin_last_year"] = np.log1p(test["sellin_last_year"])
```

### Подготовка данных для классификации факта наличия продаж

- `feat_cols` — список признаков, которые будут использоваться для обучения модели классификации.
  Включают временные характеристики и отложенные продажи, в том числе логарифмированные.
- `train_clf` — обучающая выборка, в которой известны значения целевой переменной `BPV`.
- `sale_flag` — бинарная целевая переменная, равная 1, если в периоде были продажи (`BPV > 0`), и 0 — если продаж не было.
- `X_clf` — матрица признаков для классификации.
- `y_clf` — целевой вектор наличия/отсутствия продаж.

```python
feat_cols = [
    "Period_ord", "weekofyear", "month",
    "sellin_last_year", "log_sellin_last_year",
    "sin_month", "cos_month"
]

train_clf = train[train["BPV"].notna()].copy()
train_clf["sale_flag"] = (train_clf["BPV"] > 0).astype(int)

X_clf = train_clf[feat_cols]
y_clf = train_clf["sale_flag"]
```

### Подготовка данных для регрессии и тестирования ансамбля моделей

- `train_reg` — подмножество обучающей выборки, содержащее только периоды с ненулевыми продажами (`sale_flag == 1`). Это данные для обучения регрессионной модели.
- `X_reg` — признаки для регрессии.
- `y_reg` — логарифмированные значения `BPV`, чтобы стабилизировать разброс значений и улучшить обучение модели.
- `X_test` — матрица признаков из тестовой выборки для дальнейшего прогнозирования.
- `true_test` — таблица с фактическими значениями `BPV_true` и соответствующими датами из тестовой выборки.
- `N_MODELS = 3` — количество моделей в ансамбле (например, при бутстраппинге или бэггинге).
- `probas` — список, в который будут сохраняться предсказанные вероятности наличия продаж (из классификатора).
- `preds` — список для хранения предсказаний объёма продаж (из регрессора).

```python
train_reg = train_clf[train_clf["sale_flag"] == 1].copy()
X_reg = train_reg[feat_cols]
y_reg = np.log1p(train_reg["BPV"])

X_test = test[feat_cols].reset_index(drop=True)
true_test = test[["Period", "BPV_true"]].reset_index(drop=True)

N_MODELS = 3
probas = []
preds = []
```

### Обучение ансамбля из классификаторов и регрессоров (XGBoost)

Цикл по `N_MODELS` с разными значениями `random_state` (от 42 до 44) реализует ансамблирование моделей:

1. **Классификатор (`XGBClassifier`)**:
   - Предсказывает вероятность того, что в периоде будут продажи.
   - Использует `scale_pos_weight` для компенсации дисбаланса классов (преобладание нулевых продаж).
   - Результат (`proba_test`) — массив вероятностей продажи (второй столбец предсказаний `predict_proba`), сохраняется в `probas`.

2. **Регрессор (`XGBRegressor`)**:
   - Предсказывает объём продаж `BPV`, но только на основе периодов с положительным спросом.
   - Предсказания логарифмированы во время обучения (`y_reg`) и переводятся обратно через `expm1`.
   - Отрицательные значения в прогнозах принудительно обнуляются.
   - Результаты сохраняются в список `preds`.

Таким образом, ансамбль сначала определяет, будет ли продажа, а затем оценивает объём, если продажа вероятна.

```python
for seed in range(42, 42 + N_MODELS):
    clf = XGBClassifier(
        n_estimators=500,
        learning_rate=0.1,
        max_depth=6,
        scale_pos_weight=(y_clf == 0).sum() / (y_clf == 1).sum(),
        use_label_encoder=False,
        eval_metric="logloss",
        random_state=seed,
        verbosity=0
    )
    clf.fit(X_clf, y_clf)
    proba_test = clf.predict_proba(X_test)[:, 1]
    probas.append(proba_test)

    reg = XGBRegressor(
        n_estimators=600,
        learning_rate=0.05,
        max_depth=6,
        random_state=seed,
        verbosity=0
    )
    reg.fit(X_reg, y_reg)
    pred = np.expm1(reg.predict(X_test))
    preds.append(np.maximum(pred, 0))
```

### Комбинирование результатов ансамбля и оценка финального прогноза

- `proba_test_median` — медианное значение вероятностей продажи, полученное от всех классификаторов.
- `base_pred_median` — медианный прогноз объёма продаж, полученный от всех регрессоров.
- `threshold = 0.05` — порог вероятности, выше которого считается, что продажа состоится.
- `is_sale` — бинарный массив, в котором `True` указывает на наличие продажи (вероятность выше порога).
- `y_pred` — итоговый прогноз:
  - если вероятность продажи больше порога, используется медианный прогноз объёма;
  - иначе — прогноз равен нулю.

- `merged` — объединённый датафрейм с датой (`Period`), реальным (`BPV_true`) и предсказанным (`BPV_pred`) значениями.

- Расчёт метрик качества прогноза:
  - `WAPE` — взвешенная средняя абсолютная ошибка,
  - `MAE` — средняя абсолютная ошибка,
  - `RMSE` — среднеквадратичная ошибка,
  - `MASE` — масштабированная ошибка (при `m=1`, что соответствует недельной сезонности).

Метрики выводятся для анализа точности работы комбинированной модели.

```python
proba_test_median = np.median(probas, axis=0)
base_pred_median = np.median(preds, axis=0)

threshold = 0.05
is_sale = proba_test_median >= threshold
y_pred = np.where(is_sale, base_pred_median, 0)

merged = pd.DataFrame({
    "Period": true_test["Period"],
    "BPV_true": true_test["BPV_true"],
    "BPV_pred": y_pred
})

metrics = compute_metrics(merged["BPV_true"], merged["BPV_pred"], m=1)

print(f"WAPE = {metrics['WAPE']:.2f}%")
print(f"MAE  = {metrics['MAE']:.4f}")
print(f"RMSE = {metrics['RMSE']:.4f}")
print(f"MASE = {metrics['MASE']:.4f}")
```
